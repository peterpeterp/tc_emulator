```python
import os, glob, sys,datetime

import xarray as xr
import numpy as np

os.chdir('/home/peterp/projects/data/SST/OISST/raw')
```

```python

for year in range(2023,2025):
    for month in range(5,11):
        for day in range(1,32):
            link_root = f"https://www.ncei.noaa.gov/data/sea-surface-temperature-optimum-interpolation/v2.1/access/avhrr/{year}{str(month).zfill(2)}/"
            fl = f"oisst-avhrr-v02r01.{year}{str(month).zfill(2)}{str(day).zfill(2)}.nc"
            os.system(f"wget {link_root}{fl}")
    sst = xr.open_mfdataset(f"oisst-avhrr-v02r01.{year}*.nc").squeeze()['sst'].loc[:, 10:20, 360-90:360-20]
    weights = np.cos(np.deg2rad(sst.lat))
    sst_mdr = sst.weighted(weights).mean(("lon", "lat"))
    xr.Dataset({'sst':sst_mdr}).to_netcdf(f'../sst_mdr_{year}.nc')
```
